```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

x,v = np.loadtxt('droptower_vdata.txt', unpack=True)
height = integrate.cumtrapz(v,x,initial=0)
a = np.diff(v)

plt.plot(x, height, label='h')
plt.plot(x, v,label='v')
plt.plot(x[1:], a, label='a')
plt.legend()

a_up = a[a > 0]
avg = np.average(a_up)
print(avg)
```
